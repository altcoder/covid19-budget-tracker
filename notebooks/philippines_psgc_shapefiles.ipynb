```python
# papermill parameters
output_folder = '../output/'
execution_date = '2020-03-28'
execution_time = '2020-03-28T00:00:00'
```

```python
import requests
import zipfile
import geopandas
import os
```

```python
# Download Municipalities Dataset
datasets_folder = "%s/../datasets/" % output_folder
url = 'https://media.githubusercontent.com/media/altcoder/philippines-psgc-shapefiles/master/datasets/SHAPEFILES/MunicipalitiesMin.zip'
r = requests.get(url)  

shp_zip = '{}/MunicipalitiesMin.zip'.format(datasets_folder)
with open(shp_zip, 'wb') as f:
    f.write(r.content)

with zipfile.ZipFile(shp_zip, 'r') as f:
    f.extractall(datasets_folder)

os.remove(shp_zip)
```

```python
# Read shapefile
adm3 = geopandas.read_file('{}/MunicipalitiesMin.shp'.format(datasets_folder))
adm3 = adm3[['ADM_ID','ADM1_PCODE', 'ADM1_EN', 'ADM2_PCODE', 'ADM2_EN', 'ADM3_PCODE', 'ADM3_EN', 'LATITUDE', 'LONGITUDE', 'AREA', 'PERIMETER']]
adm3 = adm3.drop_duplicates() # Entry is per geometry so make sure to dedup

adm3_csv = '{}/psgc_adm3.csv'.format(datasets_folder)
adm3.to_csv(adm3_csv, index=False)
adm3
```
